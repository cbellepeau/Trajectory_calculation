```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

x = [50, 40, 70, 60, 55]
y = [50, 40, 40, 60, 55]
label = ['target', 'january', 'february', 'march', 'april']

# Spécifiez une liste de couleurs que vous souhaitez utiliser pour chaque point
colors = ['red', 'blue', 'green', 'purple', 'orange']

# Créer le nuage de points avec des couleurs différentes
plt.scatter(x, y, c=colors)

# Ajouter des labels pour chaque point
for i in range(len(x)):
    plt.annotate(label[i], (x[i], y[i]), textcoords="offset points", xytext=(0, 10), ha='center')

plt.xlim(30, 80)
plt.ylim(30, 70)

# Ajouter des labels pour les axes
plt.xlabel('mouse sales (k€)')
plt.ylabel('keyboard sales (k€)')

# Afficher le graphe
plt.title('Tech store sales')
plt.show()
```
